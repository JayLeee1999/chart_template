```python
import matplotlib.pylab as plt
# '-' 깨짐 해결 모듈
import matplotlib as mpl

plt.rcParams["font.family"] = "Malgun Gothic"  # For Windows
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (8,4)
print(plt.rcParams["font.family"])

# 마이너스 깨짐 해결
mpl.rcParams["axes.unicode_minus"] = False
```

# 밑 그림 그리기

```python
data1 = np.arange(1,50)
data1
```

```python
data2 = np.arange(51,101)
data2
```

```python
data3 = np.arange(101,150)
data3
```

# 다중 그래프
- 하나의 figure 에 두개 이상의 데이터를 표현함

```python
plt.plot(data1)
plt.plot(data2)
plt.plot(data3)
plt.show()
```

## subplot 그리기
- subplot(row, column, index)
- 여러개의 그래프를 배열 형태로 그리기

```python
plt.rcParams['figure.figsize'] = (5, 5)
data1 = np.arange(100,201)
data2 = np.arange(200,301)

plt.subplot(2,1,1)
plt.plot(data1)
plt.subplot(2,1,2)
plt.plot(data2)

plt.show()
```

# 여러개의 plot을 그리는 방법
-행, 열 모두 여러개로 그리기

```python
plt.rcParams["figure.figsize"] = (16, 8)
plt.rcParams["font.size"] = 12

data = np.arange(1, 51)
# data 생성

# 밑 그림
# fig, axes = plt.subplots(그래프행수, 그래프열수)
fig, axes = plt.subplots(2, 3)
# _, axes = plt.subplots(2, 3)
# axes = plt.subplots(2,3)
# fig.legend

axes[0,0].plot(data*data)
axes[1,2].plot(data**3)
axes[1,0].plot(data%10)
axes[1,1].plot(-data)
axes[0,1].plot(data//20)

# subplot 간의 간격조절
plt.tight_layout(pad=5.5)
plt.title('그래프')
```
