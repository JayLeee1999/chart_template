```python
# 데이터
regions = ['경기도', '서울', '충청도', '경상도', '전라도', '강원도', '제주도']
population = [13405000, 9720846, 2099266, 3416900, 1798980, 1541502, 670989]
```

```python
plt.bar(regions, population)
plt.title('지역 별 인구 수')
plt.xlabel('지역')
plt.ylabel('인구')
plt.show()
```

```python
df = pd.DataFrame(zip(regions, population), columns=['지역', '인구수'])
df
```

```python
df_sort = df.sort_values('인구수', ascending=False)
df_sort
```

```python
plt.bar(df_sort['지역'], df_sort['인구수'])
# 그래프 제목과 축 레이블 설정
plt.title('대한민구 지역별 인구 분포')
plt.xlabel('지역')
plt.ylabel('인구수')

# 그래프 보여주기
plt.show()
```

```python
# 박스플롯

import numpy as np

np.random.seed(0) # 매번 같은 값을 출력
male_heights = np.random.normal(loc=175, scale=7, size=100)
female_heights = np.random.normal(loc=162, scale=6, size=100)
plt.boxplot([male_heights, female_heights], labels=['male','female'])
plt.title('성별에 따른 키 분포도')
plt.xlabel('성별')
plt.ylabel('cm')
```
